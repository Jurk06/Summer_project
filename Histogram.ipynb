# Histogram
*  Approximate representation of the distribution of the numerical data
*  To construct Histogram 
1.  Bin the range values- 
(Bin -Divide the entire range into small series of intervals, and then count how many values fall into each interva )
*  e.g-if you have data about a group of people, you might want to arrange their ages into a smaller number of age intervals (for example, grouping every five years together.
*  If the bins are of equal size, a rectangle is erected over the bin with height proportional to the frequency-the number of cases in each bin.
*  Histograms give a rough sense of the density of the underlying distribution of the data
*  stimating the probability density function of the underlying variable
*  A histogram is used for continuous data, where the bins represent ranges of data
